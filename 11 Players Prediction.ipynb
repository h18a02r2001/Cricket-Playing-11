### Creating the New Dataset of Players_1 and their Strike rate

For finding the strike rate first we have to find out the total runs score by each player in column "Players_1". 

Fornula for finnding the Strike rate: (Number of runs/ Total Ball Played)*100

```python
# Finding out the total runs scored by the Players of column "Players_1" and "Players_2"
MergeData['Player1_Trun']= 4*MergeData['4s_1'] + 6*MergeData['6s_1']+ MergeData['Bat_Runs_1']
MergeData['Player2_Trun']= 4*MergeData['4s_2'] + 6*MergeData['6s_2']+ MergeData['Bat_Runs_2']
```

```python
# Calculating the Strike rate of the players from both the columns
MergeData['Player1_Strike']= (MergeData['Player1_Trun'] / MergeData['Ball Faced_1'])*100
MergeData['Player2_Strike']= (MergeData['Player2_Trun'] / MergeData['Ball Faced_2'])*100
```

```python
MergeData.head()
```

```python
# Creating New dataset for Player 1 Player
Player1 = MergeData[['Player1_Strike','Players_1']]

cols = ['Player1_Strike','Players_1']

Player1Strike = Player1[cols]
Player1Strike.head(10)
```

```python
# Dropping NA values
Player1Strike = Player1Strike.dropna()
```

```python
# Converting decimal numbers into numbers of Strike rate
Player1Strike = Player1Strike.astype({'Player1_Strike': 'int'})
```

```python
Player1Strike.head()
```

```python
# Changing the Columns name in Player1Strike dataset
Player1Strike.rename(columns={'Player1_Strike': 'Strike_Rate', 'Players_1': 'Player'}, inplace=True)
```
